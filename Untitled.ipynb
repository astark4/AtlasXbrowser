```r
  Load_AtlasXomics <- function(
    data.dir,
    assay = 'Spatial',
    slice = 'slice1',
    filter.matrix = TRUE,
    to.upper = FALSE,
    ...
  ) {
    if (length(x = data.dir) > 1) {
      warning("'Load_AtlasXomics' accepts only one 'data.dir'", immediate. = TRUE)   
      data.dir <- data.dir[1]
    }
    data <- Read10X(data.dir = data.dir, ...)
    if (to.upper) {
      rownames(x = data) <- toupper(x = rownames(x = data))
    }
    object <- CreateSeuratObject(counts = data, assay = assay)
    starting_wd = getwd()
    setwd(dir = file.path(data.dir, 'spatial'))
    png_file <- list.files(pattern = "\\.png$")
    if (length(png_file) == 1) {
    file.copy(png_file,'tissue_hires_image.png')
    file.copy(png_file,'tissue_lowres_image.png')
    }
    setwd(starting_wd)
    image <- Read10X_Image(
      image.dir = file.path(data.dir, 'spatial'),
      filter.matrix = filter.matrix
    )
    image <- image[Cells(x = object)]
    DefaultAssay(object = image) <- assay
    object[[slice]] <- image
    return(object)
    }
```

```r
object_AXOSpatial_seurat = Load_AtlasXomics(data.dir = dataset)

mat = as.data.frame(object_AXOSpatial_seurat@assays$Spatial@counts)

mat2 = mat[, colSums(mat != 0) > 0]

object_AXOSpatial_seurat = subset(object_AXOSpatial_seurat, cells = colnames(mat2))


VlnPlot(object_AXOSpatial_seurat, features = "nCount_Spatial", pt.size = 0.1) + NoLegend()
SpatialFeaturePlot(object_AXOSpatial_seurat, features = "nCount_Spatial", pt.size.factor=100) + ggplot2::theme(legend.position = "right")
```

```r
object_AXOSpatial_seurat[["percent.mt"]] <- PercentageFeatureSet(object_AXOSpatial_seurat, pattern = "^mt-")

VlnPlot(object_AXOSpatial_seurat, features = "percent.mt", pt.size = 0.1) + NoLegend()
SpatialFeaturePlot(object_AXOSpatial_seurat, features = "percent.mt",pt.size.factor = 100.8, alpha = c(0.5, 1)) + ggplot2::theme(legend.position = "right")

object_AXOSpatial_seurat <- SCTransform(object_AXOSpatial_seurat, assay = "Spatial", verbose = FALSE)
```

```r
object_AXOSpatial_seurat <- RunPCA(object_AXOSpatial_seurat, assay = "SCT", verbose = FALSE)
object_AXOSpatial_seurat <- FindNeighbors(object_AXOSpatial_seurat, reduction = "pca", dims = 1:30)
object_AXOSpatial_seurat <- FindClusters(object_AXOSpatial_seurat, verbose = FALSE)
object_AXOSpatial_seurat <- RunUMAP(object_AXOSpatial_seurat, reduction = "pca", dims = 1:30)

# Idents(object_AXOSpatial_seurat) = 'SCT_snn_res.0.8'

DimPlot(object_AXOSpatial_seurat, reduction = "umap", label = TRUE,pt.size = 2)
SpatialDimPlot(object_AXOSpatial_seurat, label = TRUE, label.size = 3, pt.size.factor = 100.8)
```
